```python
# Exploratory Analysis of Conversion Rates

This notebook explores user counts and conversion rates for control and treatment groups.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/ab_corrected.csv",sep = ";")
```

```python
summary = (
    df.groupby("group")
      .agg(
          users=("user_id", "count"),
          conversions=("converted", "sum"),
          conversion_rate=("converted", "mean")
      )
)

summary
```

```python
summary['conversion_rate'].plot(kind='bar')
plt.ylabel("Conversion Rate")
plt.title("Conversion Rate by Experiment Group")
plt.show()
```

```python
The control group shows a slightly higher conversion rate than the treatment group.
However, this observed difference may be due to random variation and requires
statistical testing.
```
